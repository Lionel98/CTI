```python
import pandas as pd

df=messages=dataset
tec_name_list = []
tec_name_unique = []
for i in df['train']['tec_name']:
 tec_name_list.append(i)


tec_name_unique = []
df_label= pd.DataFrame(tec_name_list)
for i in df_label[0].unique():
  tec_name_unique.append(i)

  df_label_unqiue= pd.DataFrame(tec_name_unique)
df_label_unqiue[0]

id2label = {idx:df_label_unqiue[0][idx] for idx in range(546)}
id2label
```

```python
label2id = {v:k for k,v in id2label.items()}  
label2id
```

```python
labels_df = ds["train"].to_pandas()
labels_df["tec_name"].value_counts(normalize=True).sort_index()
```

```python
from transformers import AutoTokenizer
model_ckpt = "ltkw98/minilm-finetuned_3"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
tokenizer(ds["train"]["sentence"][:1])
```

```python
def tokenize_text(examples):
  return tokenizer(examples["sentence"], truncation=True, max_length=512)
```

```python
ds = ds.map(tokenize_text, batched=True)
ds
```

```python
ds.push_to_hub("ltkw98/Tokenize_Dataset1")
```

```python
class_weights = (1 - (labels_df["label"].value_counts().sort_index() / len(labels_df))).values
class_weights
```

```python
import torch
class_weights = torch.from_numpy(class_weights).float().to("cuda")
class_weights
```

```python
ds = ds.rename_column("label", "labels")
```

```python
from torch import nn
import torch
from transformers import Trainer

class WeightedLossTrainer(Trainer):
  def compute_loss(self, model, inputs, return_outputs=False):
    outputs = model(**inputs)
    logits = outputs.get("logits")
    labels = inputs.get("labels")
    loss_func = nn.CrossEntropyLoss(weight=class_weights)
    loss = loss_func(logits, labels)
    return (loss, outputs) if return_outputs else loss
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model_ckpt, num_labels=546, id2label=id2label, label2id=label2id)
```

```python
from sklearn.metrics import f1_score

def compute_metrics(pred):
  labels = pred.label_ids
  preds = pred.predictions.argmax(-1)
  f1 = f1_score(labels, preds, average="weighted")
  return {"f1": f1}
```

```python
from transformers import TrainingArguments
batch_size = 64
logging_steps = len(ds["train"])
output_dir1="minilm-finetuned_4"

training_args = TrainingArguments(output_dir=output_dir1,
                                  num_train_epochs=100,
                                  learning_rate=2e-5,
                                  per_device_train_batch_size=4,
                                  per_device_eval_batch_size=batch_size,
                                  gradient_accumulation_steps=16,
                                  weight_decay=0.01,
                                  evaluation_strategy="epoch",
                                  push_to_hub=True)
```

```python
trainer = WeightedLossTrainer(model=model,
                              args=training_args,
                              compute_metrics=compute_metrics,
                              train_dataset=ds["train"],
                              eval_dataset=ds["validation"],
                              tokenizer=tokenizer)
```

```python
trainer.train()
```
