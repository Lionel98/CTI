### **FOLD0**

```python
from datasets import load_dataset
from datasets import get_dataset_split_names
test = load_dataset("csv", data_files="/content/test_set.csv")
test
```

```python
validation_fold0 = load_dataset("csv", data_files="/content/validation_fold0.csv")
validation_fold0
```
